```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_column', None)
pd.set_option('display.max_colwidth', None)


def unique_data(dataSales):
  data_unique = pd.DataFrame()
  data_unique['Columns']= dataSales.columns
  data_unique['Unique Count'] = [dataSales[col].nunique() for col in (dataSales.columns)]
  data_unique['Unique'] = [dataSales[col].unique() for col in (dataSales.columns)]
  return data_unique

unique_data(dataSales)
```
